```python
def getAndPreprocess(name):
    img = cv.imread(f"training_images/{name}")[:, :, 0]
    img = np.invert(np.array([img]))
    return img[0]

def showImage(img):
    plt.imshow(img, cmap=plt.cm.binary)
    plt.show()
```
